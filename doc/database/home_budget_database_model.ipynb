# __Relational database model for *"home budget app"*.__
***

# **A short story abouth the database**

***
## **Budget**

The **budget** is set up for one month.
One user can have several budgets (January 2021, February 2021, March 2021, etc...) as well as a single budget can have several users. We can share it with someone.

***
## **Income**

**Income** is an important item in the budget. Each income has a **name**, **amount**, **date**, and status that determines whether it is **steady_income** or not. A budget can have many sources of funding. We must deposit all income. One income can be **located** to several **deposits**.

***
## **Deposit**

The **deposit** can be my account, sub-account, wife's account, lover's account or a sock made by a beloved grandmother. The deposit has a **name** and a total **amount**. One deposit can be funded from several sources.
Several purposes for which we **save** can be assigned to the deposit. We can pay for **debts** and **expenses** from the deposit.

***
## **Expense**

Every budget has its expenses. Each **expense** has its own **name**, e.g. *rent cost* or *new TV set* and we can assign it to a specific **category** (*rent* -> *fixed expenses*, *new TV* -> *whims*). 
It also has a status that determines whether it is a **steady_expense**. The expense also has a **date_of_payment**, we also need to know where to get the money to settle the expense, so we reserve it on a specific **deposit**. 
We can pay the expence at once or in several **payments**.

***
## **Debt**

The budget may be burdened with debts. Each **debt** has its **name**, **date_of_payment** and **amount**. We also need to specify the **deposit** from which we will pay it. We can pay for it one time or in several **payments**.

***
## **Saving**

It's good to have savings.
It is good to **name** the purpose for which we save money and the **deposit** where we will keep our savings. It is also good to know the saved **amount**. We can put money on the saving deposit many times.
